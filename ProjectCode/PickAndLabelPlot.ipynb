```python
%matplotlib inline
%matplotlib nbagg
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.DataFrame(data={"x":[1,2,3,4], "y":[10,20,30,40]}, index=["OR", "WA", "CA", "ID"])
```

```python
def onpick(g, event):
    ind = event.ind
    g.set_title(df.index[ind[0]])
    return

fig = plt.figure()
g = fig.add_subplot(111)

g.scatter(df.x, df.y, picker=True)
fig.canvas.mpl_connect('pick_event', lambda e: onpick(g, e))
```
